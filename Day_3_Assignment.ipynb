Question 1:
Plot using sns.relplot, line plot, box plot on fmri data set



```python
import seaborn as sns #  Python data visualization library based on matplotlib for statistical graphics
import pandas as pd # data analysis and manipulation tool
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

**relplot()**

```python
sns.relplot(x="timepoint", y="signal", col="region",hue="event", style="event",kind="line", data=fmri)
```

**lineplot()**

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
```

**boxplot()**

```python
sns.boxplot(x="timepoint",y="signal", hue ="region", data=fmri)
```
